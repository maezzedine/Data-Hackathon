## Normalizing the field Job

```python
df["Job"].unique().tolist()
```

```python
jobs = [ "Human resources", "Accountant",  "Secretary", "Entrepreneur", "Business development", "Consulting", "Marketing", "Business", "Finance", "Economics",
            "Machine Learning", "Data Science", "Web Developer", "Software Engineer", "Computer Science", "Computer Communication Engineering", "banking", "Sales Management", "Retails" ]

def extract_job(job):
    for j in jobs:
        if j in job:
            return j

extracted_jobs = []
for index, row in df.iterrows():
    extracted_jobs.append(extract_job(row['Job']))

df['extracted_job'] = extracted_jobs
```

```python
df
```

## Normalizing the location field

```python
df["Location"].unique().tolist()
```

```python
import math

locations = ['Lebanon', 'United Arab Emirates', 'Italy', 'Spain', 'Saudi Arabia', 'United States', 'Congo', 'Switzerland', 'Luxembourg',
             'Romania', 'Turkey', 'Qatar', 'Canada', 'United Kingdom', 'France', 'Italy', 'Cyprus', 'Armenia']

def extract_location(loc):
    if type(loc) != str and math.isnan(loc): return
    
    for l in locations:
        if l in loc:
            return l

extracted_locations = []
locations = []
for index, row in df.iterrows():
    location = row['Location']
    
    extracted_locations.append(extract_location(location))
    if type(location) != str and math.isnan(location):
        locations.append(None)
    else:
        locations.append(location[:location.find('Contact info')])

df['extracted_location'] = extracted_locations
df['Location'] = locations
```

```python
df
```

## Normalizing University field

```python
df['College'].unique().tolist()
```

```python
universities = [ 'Lebanese University', 'Notre Dame University', 'Online Learning Consortium', 'University of Balamand', 'Lebanese International University', 'American University of Beirut', 'Haigazian University', 'Lebanese American University', 'Université Saint', 'Saint Joseph University of Beirut', 'Beirut Arab University', 'American University of Science and Technology', 'Holy Spirit University of Kaslik', 'Kingston University', 'AUL university', 'Université de Franche', 'Lebanese university faculty of agronomy', 'Antonine University']


def extract_university(uni):
    if type(uni) != str and math.isnan(uni): return
    
    for u in universities:
        if u in uni:
            return u

extracted_universities = []
for index, row in df.iterrows():
    location = row['College']
    
    extracted_universities.append(extract_university(location))

df['extracted_university'] = extracted_universities
```

```python
df
```

## Normalize Certification

```python
certifications = []
for index, row in df.iterrows():
    cert = row['Certification']
    
    if (type(cert) == str):
        cert = cert.split('\n')
        certifications.append(len(cert)-2)
    else:
        certifications.append(0)

df['extracted_certifications'] = certifications
```
